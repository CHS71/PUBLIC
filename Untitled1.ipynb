```python
import os
import re
import glob




def make_file(f_in,input_file_name,path):

    output_file_name = re.sub('[^a-z|A-Z|0-9|가-힣|\u4E00-\u9FFF|\u3040-\u309Fー|\u30A0-\u30FF|\\\]','' ,input_file_name ) 

    end_num = re.search('20..|19..',output_file_name)

    output_file_name =output_file_name[:end_num.regs[0][0]]

    start_num2 = output_file_name.rfind('\\')        

    output_file_name = path + output_file_name[start_num2:]+'.txt'


        
    if input_file_name.endswith('.srt'):
        f_out = open(output_file_name, 'w',encoding='utf8')

    else :
        f_out = open(output_file_name, 'w')



    lines = filter(None, (line.strip() for line in f_in) )


    seq_num = 1
    
    
    if input_file_name.endswith('.smi'):
        
        for line in lines:
            
            try:

                if int(line) < 9999:

                    if int(line) == 1:
                        seq_num = seq_num +1   
                    
                    if int(line) != seq_num:    
                        
                        pass
                    
                    else :
                        f_out.write('\n')  
                        seq_num = seq_num + 1   
                        
                else:
                    line = line + ' '
                    f_out.write(line)                   

            
            except:
                    
                stime_pattern = re.compile('(\d{2}:\d{2}:\d{2},)')                

                if stime_pattern.match(line) is not None:
                    continue
                
                
                word = ['♪','<>','--','-']       
                # ! , '' , . , = , "" , 
        
                for i in re.findall(r"<(.+?)>", line):

                    line = re.sub(i,'',line)
                
                
                for w in word:
                    line = re.sub(w,'',line)   
                    
                line = line + ' '
                f_out.write(line)



    f_out.close()                
                
                    
                
                



    
    
def encoding_func(input_file_name, path):
    
    
    global error_lst
    global normal_tt
    global utf8_tt
    global latin1_tt
    global error_tt
    
    
    error=''
#     print(input_file_name)    

    
    try:
        with open(input_file_name) as f_in:
            make_file(f_in,input_file_name,path)
#             print('normal')
            normal_tt=normal_tt+1
    except:
        try:
            with open(input_file_name,encoding='utf-8-sig') as f_in:
                make_file(f_in,input_file_name,path)
#                 print('utf8 is right')
                utf8_tt=utf8_tt+1
        except:                     

            print('\n[error]',input_file_name,'\n')

            error_lst.append(input_file_name)
            error_tt=error_tt+1


                

    f_in.close()
    
    return error



def main():
    
    global error_lst
    global normal_tt
    global utf8_tt
    global latin1_tt
    global error_tt
    
    
    p =r'C:\Users\CHS\Desktop\wcoding\data\영화크롤링.vol2'### 경로 설정해줄 것.
    new_path =r'C:\Users\CHS\Desktop\wcoding\data\영화크롤링.vol2_text'
    
    testt =r'C:\Users\CHS\Desktop\wcoding\data\영화크롤링.vol2\자막\[Shiniori-Raws]+009+Re+Cyborg+The+Movie+BD+1280x720+x264+AAC.srt'
    error_t = encoding_func( testt , p )  



error_lst=[]

normal_tt = 0
utf8_tt = 0
latin1_tt = 0
error_tt = 0

if __name__ == '__main__':
    main()
```
