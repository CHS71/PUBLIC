### IPython의 도움말과 문서

IPython의 가장 유용한 기능 중 하나는 모르는 정보를 효과적으로 찾도록 도움을 주는 것이다.<br>

아래와 같은 정보들을 쉽게 찾을 수 있다.
1. 함수의 호출방법,함수의 인수와 옵션
2. 객체의 소스코드
3. 패키지 정보, 객체의 속성이나 메서드

* #### ?로 문서 확인하기

IPython에서는 단축키로 ? 문자를 사용하여 모든 파이썬 객체에 대한 간결한 요약 정보와 사용법을 확인할 수 있다.<br>
파이썬에서는 help() 내장함수를 통해 확인이 가능하다. 

```python
len?
```

<br>
? 문자는 직접 만든 함수나 다른 객체에서도 동작한다.

먼저 두 입력값의 합을 반환하는 test_sum함수를 생성한다.

```python
def test_sum(a,b):
    '''a,b의 합을 반환'''
    result = a + b
    return result
```

''' a,b의 합을 반환 ''' 부분은 함수의 독스트링을 만들기 위해 첫줄에 문자열 리터럴을 삽입한 것이다.

```python
test_sum?
```

이렇게 독스트링을 통해 문서에 손쉽게 접근할 수 있기 때문에 코드를 작성할 때 항상 이러한 인라인 문서를 추가하는 습관을 가지는 것이 좋다.

* #### ??로 소스코드에 접근하기

?? 문자는 객체의 소스코드에 접근하게 해준다. 객체의 구현 세부사항을 파악함으로써 또 다른 아이디어를 얻을 수 있다. 

?? 문자를 사용하다보면 소스코드를 보여주지 않는 경우도 있다. 이는 질문한 객체가 C나 다른 컴파일된 확장 언어로 구현됐기때문이다.

<br><br>

### IPython 셸에서 사용할 수 있는 키보드 단축키

IPython 셸은 명령어를 입력하는동안 빠르게 탐색할 수 있는 여러 단축키를 제공한다.<br><br>
이러한 단축키들은 실제로 GNU Readline 라이브러리의 종속성을 통해 제공된다.<br><br>
이 중 일부는 브라우저 기반의 노트북에서는 동작하지만 이번 절에서는 주로 IPython 셸의 단축키를 다룬다.

<br><br>
* #### 탐색 단축키

    * Ctrl-a  : 커서를 줄의 맨 앞으로 옮김
    * Ctrl-e  : 커서를 줄의 맨 뒤로 옮기
    * Ctrl-b(또는 왼쪽 화살표 키)  : 커서를 한 글자 뒤로 옮김
    * Ctrl-f(또는 오른쪽 화살표 키)  : 커서를 한 글자 앞으로 옮김

<br>

* #### 텍스트 입력 단축키

    * 백스페이스 키 : 그 줄의 이전 문자 삭제
    * Ctrl-d : 그 줄의 다음 문자 삭제
    * Ctrl-k : 현재 커서 위치에서 그 줄 마지막까지 텍스트를 잘라냄
    * Ctrl-u : 그 줄의 처음부터 현재 커서 위치까지 텍스트를 잘라냄
    * Ctrl-y : 이전에 잘라낸 텍스트를 끌어당김(yank,즉 붙여넣기)
    * Ctrl-t : 앞의 두 글자의 위치를 바꿈

<br>

* #### 명령어 이력 단축키

    * Ctrl-p(또는 위쪽 화살표 키) : 이력에서 이전 명령어에 접근
    * Ctrl-n(또는 아래쪽 화살표 키) : 이력에서 다음 명령어에 접근
    * Ctrl-r : 명령어 이력을 역탐색

<br>

* #### 기타 단축키

    * Ctrl-l : 터미널 화면을 지움
    * Ctrl-c : 현재 파이썬 명령어를 중단
    * Ctrl-d : IPython 세션을 종료

<br><br>

### IPython 매직 명령어

이전 두 정에서는 IPython에서 파이썬을 대화형으로 효과적으로 사용하고 탐색하는 법을 살펴봤다.<br>
여기서는 IPython이 일반 파이썬 구문에 추가한 몇 가지 향상된 기능에 대해 살펴본다.

이것을 매직 명령어라고 하며, 맨 앞에 %기호를 붙인다.<br>
매직명령어는 일반적인 데이터 분석에서 흔히 발생하는 다양한 문제점을 간결하게 해결하도록 고안됐다.

매직명령어는 % 를 접두사로 붙이고 입력값 한 줄에 대해 동작하는 **라인 매직** 과 %% 를 붙이고 입력값 여러줄에 대해 동작하는 **셀 매직**으로 나뉜다.

<br>

* #### 코드 블록 붙여넣기: %paste와 %cpaste

IPython 인터프리터를 이용할 때 특히 들여쓰기와 인터프리터 표시자가 포함될 경우 <br>여러 줄로 된 코드 블록을 붙여넣으면 예기치 않은 오류가 쉽게 발생할 수 있다. 간단한 예를 통해 알아보자.

ipython 셸에서 함수를 생성하고 소스코드를 그대로 복사해보자.

```python
Image('C:/jupyter_project/image/image9.JPG')
```

복사한 것을 그대로 붙여넣어보면 프롬프트 문자 때문에 인터프리터 혼선이 생긴다.<br>
이때  %paste 매직 함수가 여러 줄의 인터프리터 표시자가 포함된 입력값을 처리할 수 있도록 해준다.

```python
Image('C:/jupyter_project/image/image10.JPG')
```

%cpaste 블록을 사용하면 실행 전에 코드를 마음껏 붙여 넣을 수 있다.

```python
Image('C:/jupyter_project/image/image11.JPG')
```

<br>

* #### 외부 코드 실행: %run

방대한 코드를 작성하다보면 대화형 탐색을 위한 IPython과 재사용할 코드를 저장하기 위한 텍스트 편집기를 함께 사용하게된다.<br>
이렇게 저장한 코드를 새 창에서 실행하기 보다는 IPython 세션 안에서 실행하는 것이 편리할 수 있다.

```python
def square(x):
    """숫자의 제곱을 반환"""
    return x**2

for N in range(1,4):
    print(N,"의 제곱은",square(N))
```
