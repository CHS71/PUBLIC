```python
from openpyxl import load_workbook

import pandas as pd

# DART 이중 딕셔너리 생성하기(ifrs_ , dart_)

# load_wb = load_workbook("C:/DATA/dart.xlsx",data_only = True)
# sheet_lit = load_wb.get_sheet_names()


DART = {}


# input : sheet_lit(리스트) , DART(딕셔너리)

for j in sheet_lit:
    
    sheet=load_wb[j]

    DART_max_row = sheet.max_row
    
    if j not in DART:
        DART[j] = []
        
    ifrs_=[]
    dart_=[]

    for i in range(2,DART_max_row+1):

        if not sheet.cell(row=i,column = 6).value:
            continue
            
            
        if sheet.cell(row=i,column =5).value.strip()[:4] == 'ifrs':
        
            ifrs_.append(sheet.cell(row=i,column = 6).value)
            
            
        if sheet.cell(row=i,column =5).value.strip()[:4] == 'dart':
        
            dart_.append(sheet.cell(row=i,column = 6).value)
        
    
    DART[j]={}
    DART[j]['ifrs'] = ifrs_
    DART[j]['dart'] = dart_
    
    
    
```

```python
#df_DART 와 IFRS19_roll의 비교(IFRS19_roll기준으로)


## hardcoding: case2 딕셔너리 생성  /  case2_1 , case2_2 리스트
case2={}


## input : IFRS19_roll(데이터프레임) / case2(딕셔너리) / DART(딕셔너리)

for i in IFRS19_roll:
    
    if i not in case2:
        case2[i]=[]
    
    case2_1=[]
    case2_2=[]
    
    for j in set(IFRS19_roll[i]) - set(DART[i]['ifrs']):
        
        if j in DART[i]['dart']:
            
            case2_2.append(j)
            
        else:
            
            case2_1.append(j)
    
    case2[i]={}
    case2[i]['2_1'] = case2_1
    case2[i]['2_2'] = case2_2
    
    
```

```python
#IFRS19 overall

IFRS19_overall =[]

pattern = '^\[\d+\]'

for i in range(2,test_max_row+1):
    
    if re.search(pattern,test1_1.cell(row=i,column = 1).value):
        continue
        
    IFRS19_overall.append(test1_1.cell(row=i,column = 1).value)
```

```python
# Case1 / Case3 항목 추출
## DART제무제표에서 항목이 중복되는 경우를 고려하지 않음.  -> set( DART[i]['ifrs'] ) 사용하지않음.

case={}


for i in IFRS19_roll:
    
    if i not in case2:
        case[i]=[]
    
    case1=[]
    case3=[]
    
    for j in DART[i]['ifrs']:
        
        if j in IFRS19_overall:
            
            case1.append(j)
            
        else:
            
            case3.append(j)
    
    case[i]={}
    case[i]['1'] = case1
    case[i]['3'] = case3
```
