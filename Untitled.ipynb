```python
# 자막파일의 자막을 txt로 뽑아내기 (불필요한 단어 삭제)

import glob
import os
import re

def clear_dir(path):
    print('[+] 현재 경로       : ' + path)
    file_list = glob.glob(path+'/*')
    for item in file_list:
        if item.endswith('.txt'):
            os.remove(item)

            
def make_file(f_in,input_file_name,path_len):
    output_file_name = re.sub('[^a-z|A-Z|0-9]','' ,input_file_name ) # 컴파일과 매치를 한번에 했다. 
    idx = re.search('20..|19..',output_file_name)
    output_file_name = input_file_name[:path_len+input_file_name.count('\\')+1]+output_file_name[path_len:idx.regs[0][0]]+'.txt'

    if input_file_name.endswith('.srt'):
        f_out = open(output_file_name, 'w',encoding='utf8')
    else :
        f_out = open(output_file_name, 'w')

    lines = filter(None, (line.rstrip() for line in f_in))

    for line in lines:
        
        word = ('>','자막','제공','제작','sub','Sub','SUB','<!','<','{','#','++',
                '번역','\"넷플릭스','최종수정','http','srt','수입','배급',
                '영상정보','제목','\"NETFLIX','smi','SMI','iratemotor')
        
        line = re.sub('<br>|-| *\\t |<!|vvt2srt|by|</font>|[0-9]{8}|</i>','',line).lstrip() # 추가

        # <SAMI>, <HEAD>, <TITLE>, <BODY>, <SYNC Start=1124> 
        if line.startswith(word):
            continue
        #    color:white;
        if line.endswith(';') or line.endswith('}') or line.endswith('-->'):
            continue
        #</SAMI>, </BODY>
        if line.endswith('BODY>') or line.endswith('SAMI>'):
            continue
        if line.endswith('<br>') or line.endswith('\t') or line.endswith('</font>'):
            line = line[:-4]
            line = line + ' '
            f_out.write(line)
            continue

        if line.isdigit() and len(line) < 4:
            continue
        if line.startswith('0'):
            continue



        f_out.write(line + ' ')
        #f_out.write(line + '\r\n')
    f_out.close()
            
def encoding_func(input_file_name, path_len):
    error=''
    print(input_file_name)    
    try:
        with open(input_file_name) as f_in:
            make_file(f_in,input_file_name,path_len)
    except:
        try:
            with open(input_file_name,encoding='utf8') as f_in:
                make_file(f_in,input_file_name,path_len)
        except:
            print('\n[error]',input_file_name,'\n')
            error = input_file_name

    f_in.close()
    
    return error


def main():
    error_lst=[]
    p = r'C:\Users\CHS\Desktop\wcoding\data\영화크롤링.vol2'   ### 경로 설정해줄 것.
    for(path, dir, files) in os.walk(p):
        clear_dir(path)
        subtitle_list = glob.glob(path+'\\*.smi')
        subtitle_list.extend(glob.glob(path+'\\*.srt'))
        for file_name in subtitle_list:
            error_t=encoding_func(file_name,len(p)-p.count('\\')-1)
            if (error_t!=''):
                error_lst.append(error_t)
    print('[+] 전체 디렉터리 자막 추출 성공',end='\n\n')
    print('[error_list]\n',error_lst,'\n')
    for item in error_lst:
        output_file_name = re.sub('[^a-z|A-Z|0-9]','' ,item ) # 컴파일과 매치를 한번에 했다. 
        idx = re.search('20..|19..',output_file_name)
        output_file_name = item[:len(p)-p.count('\\')-1+item.count('\\')+1]+output_file_name[len(p)-p.count('\\')-1:idx.regs[0][0]]+'.txt'
        print('[삭제]',output_file_name)
        os.remove(output_file_name)
    
if __name__ == '__main__':
    main()
```
