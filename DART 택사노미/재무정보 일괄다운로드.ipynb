# Selenium

```python
from selenium import webdriver

driver = webdriver.PhantomJS('C:/phantomjs-2.1.1-windows/phantomjs-2.1.1-windows/bin/phantomjs.exe') # 윈도우

driver.get('http://dart.fss.or.kr/dsext002/main.do')
```

<br>

## css_selector : driver.find_element_by_css_selector

```python
list =[4,5,6,7]

for i in list:
    download = driver.find_element_by_css_selector(f"#listContents > div.table_list > table > tbody > tr:nth-child(1) > td:nth-child({i}) > div > a")
    print(download.click())
```

<br>

## class : driver.find_elements_by_class_name('cen_txt')

<br>

## xpath

```python
list =[4,5,6,7]

for i in list:
    download = driver.find_element_by_xpath(f"//*[@id=\"listContents\"]/div[2]/table/tbody/tr[1]/td[{i}]/div/a")
    print(download.onclick())
```

<br>

## text : driver.find_elements_by_link_text('다운로드')


```python
element = driver.find_elements_by_link_text('다운로드')


# print(element)
for i in element:
    print(i.click())
```
