```python
#하위 디렉토리 폴더들의 파일 접근
import pandas as pd
import os 
pwd ='C:\\Users\\KICPA\\Desktop\\재무정보\\IS'

case3 = ['ifrs_SharebasedPaymentTransactions',
'ifrs_DifferenceBetweenCarryingAmountOfNoncashAssetsDistributedAndCarryingAmountOfDividendsPayable',
'ifrs_PaymentsOfFinanceLeaseLiabilitiesClassifiedAsFinancingActivities',
'ifrs_DividendsReceivedClassifiedAsFinancingActivities',
'ifrs_InterestReceivedClassifiedAsFinancingActivities',
'ifrs_IncreaseDecreaseThroughChangesInAccountingPolicies',
'ifrs_IncreaseDecreaseThroughCorrectionsOfErrors']

case3 = pd.DataFrame(case3)
case3.columns = ['항목코드']

all_dict = {}

all_corp=[]

report_name = []

for path,dirs,files in os.walk(pwd):
    
    print(path)
    for file in files:
        
        
        name_idx = file.find('보고서')+ 3
        
        if file.find('연결') < 0: 
            
            if file.find('포괄') < 0:
                file_name = file[:name_idx]
            else:
                file_name = file[:name_idx]+'포괄'
                
                
        else:
            
            if file.find('포괄') < 0:
                file_name = file[:name_idx]+'_연결'
            else:
                file_name = file[:name_idx]+'포괄'+'_연결'
      

        if file_name not in all_dict:
            
#             print("##",file_name)
#             all_dict[file_name]= [] 
            all_dict[file_name] = {}

            
#         all_dict[file_name] = {}
        # 파일 로드
        n_path = os.path.join(path,file)        
        data= pd.read_csv(n_path, sep='\t',encoding='CP949')
        
        
        # case3 항목과 일치하는 인덱스에 True 값을 넣고 일치하지않으면 False값을 넣어 bool 리스트 생성
        bool_list = {}
        corporate_n = []
        
        
        for m in case3['항목코드']: 
            
            
    
            for n in data["항목코드"]:
             
                if m not in bool_list:
                    
                    bool_list[m]=[]
                    
                
                if m == n:

                    bool_list[m].append(True)

                else:
                    bool_list[m].append(False)


            
            print("******")
            print(len(bool_list[m]))      
            print(len(data))
            
            
        
            
            company_count = len(data[bool_list[m]]["회사명"].drop_duplicates()) 
            
            corporate_n.extend([i for i in data[bool_list[m]]["회사명"].drop_duplicates() if len(i) != 0 ])
            
            print('%%%%%%%%%%%%%')
            print(company_count)
            all_dict[file_name][m] = company_count
            
            
            
        report_name.append(file_name)            
        corporate_n = list(set(corporate_n))
        all_corp.append(corporate_n)
        
            

        
        # 각 키 값에 집계한 수치를 넣기
```

```python
# 완성된 딕셔너리 
df_case3 = pd.DataFrame.from_dict({'{}_{}'.format(i,j): all_dict[i][j] 
                           for i in all_dict.keys() 
                           for j in all_dict[i].keys()},
                       orient='index')
```
