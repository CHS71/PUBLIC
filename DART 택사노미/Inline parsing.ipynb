## i:context  기간 정보 딕셔너리 생성

```python
i_period ={}

stime_pattern1 = re.compile('(\d{4}-\d{2}-\d{2})')  

stime_pattern2 = re.compile('<i:startdate>\d{4}-\d{2}-\d{2}</i:startdate>|<i:enddate>\d{4}-\d{2}-\d{2}</i:enddate>')  

stime_pattern3 = re.compile('<i:instant>\d{4}-\d{2}-\d{2}</i:instant>')


import requests

from bs4 import BeautifulSoup

url ='https://www.sec.gov/ixviewer/samples/bst/out/bst-20160930.htm'  

html = requests.get(url).text  

soup = BeautifulSoup(html,'lxml')

    

for i in soup.find_all('i:context'):
    
    print(i['id'])
    
    if i['id'] not in i_period:
        
        i_period[i['id']]= []
        
        a = str(i.find_all('i:period')[0])
        
        print(a)
        
        if  stime_pattern2.search(a):
            
            for j in stime_pattern2.findall(a):
                
                i_period[i['id']].append(stime_pattern1.search(j)[0])  # period = 0 :start  1:end
            
            
        elif stime_pattern3.search(a):
        
            i_period[i['id']].append(stime_pattern1.search(a)[0])  # instant = 0
            
        else:
            print('abnormal period type occured')
        
```
