```python
#통합


from openpyxl import load_workbook

import pandas as pd

load_wb = load_workbook("C:/Users/KICPA/Desktop/dart.xlsx",data_only = True)
sheet_lit = load_wb.get_sheet_names()


#IFRS19

load_taxonom = load_workbook("C:/Users/KICPA/Desktop/IFRS_taxonomy.xlsx",data_only = True)

test1_1 = load_taxonom["full_ifrs_entry_point"]

IFRS19 = []

for i,col in enumerate(test1_1.iter_cols(min_row=1, min_col = 1 ,max_col=1)):
    for cell in col:
        IFRS19.append(cell.value)

        
        
IFRS19_r =[]      
        
for x in IFRS19:
    
    if x[0] == '[' or x == 'Concept name':
        continue

    IFRS19_r.append(x)

 


#DART


for i in sheet_lit:
    print('@@@',i)
    test = load_wb[i]
    
    # element_id , element_name
    
    col_n = ['id','name']
    element_id = []
    element_name = []
    for i,col in enumerate(test.iter_cols(min_row=1, min_col = 5 ,max_col=6)):

        if col_n[i] == 'id':

            for cell in col:

                if not cell.value or cell.value =='Element ID' :
                    continue                
                element_id.append(cell.value)           


        else:

            for cell in col:

                if not cell.value or cell.value =='Element Name' :
                    continue
                element_name.append(cell.value)    
    
    
    s = set(element_name)#517
    
    
    case1 = [x for x in s if x in IFRS19_r]
    case1_num = len(case1)
    print('case1_num:',case1_num)
    
    
    case3 = [x for x in s if x not in IFRS19_r]

    case3 = s - set(IFRS19_r)         # 421
    case3_num = len(case3)
    print('case3_num:',case3_num)
    
    
    case2 = [x for x in IFRS19_r if x not in s]
    case2_num = len(case2)
    print('case2_num:',case2_num) 
```
